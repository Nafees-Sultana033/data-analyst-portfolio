```python
import random     #random module ise used to generate random numbers

def generate_otp():
    otp = random.randint(100000, 999999)  #the randint() function generates random integer between specified range
    return otp
#I learned how to use random to generate the values like OTPs using a function and return it.
```

```python
def send_otp_to_email(email, otp):  #Functions accpets parameters to perform operations
   print(f"OTP sent to {email}: {otp}")
#I learned how to pass values to a function and use f-strings
```

```python
def received_otp(generated_otp, user_otp):
    return user_otp == generated_otp # the comparison operator == checks whether the otp matches the generated otp
#I learned how to compare two values and return true or false
```

```python
email = input("enter your email: ") #input() function takes user input as a string
generated_otp = generate_otp()      #functions are called to execute the code inside them and return values
send_otp_to_email(email, generated_otp)
#I learned how to accept user input and how functions work together to generate and send OTPs.
```

```python
attempts = 0      #variables are used to store values that can change during program execution

while attempts < 3:        #while loop executes the code as long as the condition is true
  user_otp = input("Enter the otp: ")

  if received_otp(generated_otp, int(user_otp)): #the if statement executes code, verifies OTP by converting user input to integer and compares it
    print("otp verified successfully")              #the print function displays output
    break         #break statement is used to exit a loop immediately
  else:            #else block executes when if condition is false
    print("otp verification failed")
    attempts += 1         #the increment operator increases the value of a variable by one

if attempts == 3:         #conditional statements execute code based on comparison
    print("access denied")  #denies access after three incorrect OTP attempts

#I learned how to use variables to control OTP attempts and how loop conditions help limit those attempts.
#I understood how to take user input and why data type conversion is necessary for correct comparisons.
#I also learned how to display success and failure messages, stop OTP verification when the correct OTP is entered,
#count failed attempts, and restrict access after multiple incorrect attempts.
```
